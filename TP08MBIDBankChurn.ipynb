```python
#Importamos módulos necesarios
import pandas as pd
import io
import requests
import seaborn as sns
import timeit
import matplotlib.pyplot as plt
```

```python
#Cargamos el dataset con los datos de los clientes
url="https://raw.githubusercontent.com/Jgonco/Visualizacion_de_datos/main/BankCustomerChurnPrediction.csv"
b=requests.get(url).content
bankchurn=pd.read_csv(io.StringIO(b.decode('utf-8')))
```

```python
# Se crea el gráfico de distribución de salarios estimados por país

sns.boxplot(data=bankchurn, x='country', y='estimated_salary')
plt.title('Distribución de Salarios Estimados por país')
plt.xlabel('País')
plt.ylabel('Salario Estimado')

# Se muestra el gráfico
plt.show()
```

```python
# Se crea el gráfico de distribución de salarios estimados por país y abandono

sns.boxplot(data=bankchurn, x='country', y='estimated_salary',hue='churn')
plt.title('Distribución de Salarios Estimados por país')
plt.xlabel('País')
plt.ylabel('Salario Estimado')

# Se muestra el gráfico
plt.legend(loc='upper right')
plt.show()
```

```python
# Se crea el gráfico de distribución de saldo por país
sns.boxplot(data=bankchurn, x='country', y='balance')
plt.title('Distribución del Saldo financiero por País')
plt.xlabel('País')
plt.ylabel('Saldo financiero')


# Muestra el gráfico
plt.show()
```

```python

# Se crea el gráfico de distribución de saldo financiero por país y abandono
sns.boxplot(data=bankchurn, x='country', y='balance', hue='churn')
plt.title('Distribución del Saldo financiero por país')
plt.xlabel('País')
plt.ylabel('Saldo financiero')

# Muestra el gráfico
plt.show()
```

```python
# Se crea el gráfico de distribución de Puntuación de crédito por país
sns.boxplot(data=bankchurn, x='country', y='credit_score')
plt.title('Distribución de Puntuación de crédito por País')
plt.xlabel('País')
plt.ylabel('Puntuación de crédito')


# Muestra el gráfico
plt.show()
```

```python
# Se crea el gráfico de distribución de Puntuación de crédito por país
sns.boxplot(data=bankchurn, x='country', y='credit_score', hue='churn')
plt.title('Distribución del Puntuación de crédito por País')
plt.xlabel('País')
plt.ylabel('Puntuación de crédito')


# Muestra el gráfico
plt.show()
```

```python
# Se crea el gráfico de distribución de Edad por país y abandono
sns.boxplot(data=bankchurn, x='country', y='age', hue='churn')
plt.title('Distribución del Edad por País')
plt.xlabel('País')
plt.ylabel('Edad')


# Muestra el gráfico
plt.show()
```

```python
#Distribución cruzada de Edad y Saldo financiero
sns.jointplot(x='balance',hue='churn',y='age',data=bankchurn)
plt.title('Relación entre saldo financiero y edad', y=1.2)
plt.xlabel('Saldo financiero')
plt.ylabel('Edad')
# Muestra el gráfico
plt.legend(loc='upper right')
plt.show()
```

```python
#Consulta con condiciones
bankGermany=bankchurn[
 (bankchurn.country == 'Germany')]
#Distribución cruzada de Edad y Saldo financiero
sns.jointplot(x='balance',hue='churn',y='age',data=bankGermany)
plt.title('Relación entre saldo financiero y edad en Alemania', y=1.2)
plt.xlabel('Saldo financiero')
plt.ylabel('Edad')
# Muestra el gráfico
plt.legend(loc='upper right')
plt.show()
```
